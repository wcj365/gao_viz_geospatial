```python
import geopandas as gpd
import plotly.express as px 
```

```python
# Load the shape file of Maryland counties as a GeoDataFrame:

gdf = gpd.read_file('../data/maryland_counties.zip')
print(gdf.shape)
gdf.head()
```

```python
gdf = gdf.to_crs("WGS84")

fig = px.choropleth_mapbox(
    data_frame = gdf.set_index("OBJECTID"), # Using the id as index of the data
    geojson = gdf.geometry,                # The geometry
    locations = gdf.index,                 # The index of the data
    color = "Shape_Area",
    mapbox_style = 'open-street-map',
    center = dict(lat = 39.0457530, lon = -76.641273),
    width=1400,
    height=800,
    hover_data=["COUNTY"],
    zoom = 7,
    title="Maryland Counties"
)

#fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```

```python
fig.write_html("../results/maryland_counties.html")
fig.write_html("../results/maryland_counties_nojs.html", include_plotlyjs="cdn")
fig.write_html("../results/maryland_counties_div_only.html", full_html=False)
```
